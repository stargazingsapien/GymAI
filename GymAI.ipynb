```python
#knees calculation added
import cv2 as cv
import numpy as np
import mediapipe as mp

# Initialize MediaPipe Pose
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

# Calculate angle between 3 points
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    
    if angle > 180.0:
        angle = 360 - angle
    return angle

# Update reps (presses) and curls based on angle
def update_reps_and_curls(angle, stage, counter, curl_counter, curl_stage, exercise_type):
    # Logic for counting presses (reps)
    if exercise_type == 'press':
        if angle > 160:  # Fully extended arm for presses
            stage = "down"
        if angle < 30 and stage == "down":  # Fully contracted arm for presses
            stage = "up"
            counter += 1

    # Logic for counting curls (using left arm angle)
    if exercise_type == 'curl':
        if angle < 30:  # Arm fully contracted for curls
            curl_stage = "up"
        if angle > 140 and curl_stage == "up":  # Arm fully extended for curls
            curl_stage = "down"
            curl_counter += 1
    
    return stage, counter, curl_counter, curl_stage

# Initialize camera and counters
cap = cv.VideoCapture(0)
counter = 0  # Counter for presses
curl_counter = 0  # Counter for curls
stage = None
curl_stage = None
exercise_type = None  # Default exercise type is None

# Display exercise options on the camera feed
def display_exercise_options(image):
    cv.putText(image, 'Select Exercise: ', (50, 50), cv.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv.LINE_AA)
    cv.putText(image, 'Curls [1]  Presses [2]', (50, 90), cv.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv.LINE_AA)

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        image = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
        results = pose.process(image)
        image = cv.cvtColor(image, cv.COLOR_RGB2BGR)

        # If exercise type has not been selected, show the options on the screen
        if exercise_type is None:
            display_exercise_options(image)
        else:
            try:
                if results.pose_landmarks:
                    landmarks = results.pose_landmarks.landmark
                
                    # Left arm landmarks
                    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
                    left_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
                    left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
                
                    # Right arm landmarks
                    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
                    right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
                    right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

                    # Left leg landmarks (Hip, Knee, Ankle)
                    left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x, landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
                    left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
                    left_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
                    
                    # Right leg landmarks (Hip, Knee, Ankle)
                    right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
                    right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
                    right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]

                    # Calculate angles for arms
                    left_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
                    right_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)

                    # Calculate angles for knees
                    left_knee_angle = calculate_angle(left_hip, left_knee, left_ankle)
                    right_knee_angle = calculate_angle(right_hip, right_knee, right_ankle)

                    # Display the angle on the image
                    cv.putText(image, f'Left Arm: {left_angle:.2f}', tuple(np.multiply(left_elbow, [640, 480]).astype(int)), cv.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv.LINE_AA)
                    cv.putText(image, f'Right Arm: {right_angle:.2f}', tuple(np.multiply(right_elbow, [640, 480]).astype(int)), cv.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv.LINE_AA)

                    # Display knee angles
                    cv.putText(image, f'Left Knee: {left_knee_angle:.2f}', tuple(np.multiply(left_knee, [640, 480]).astype(int)), cv.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv.LINE_AA)
                    cv.putText(image, f'Right Knee: {right_knee_angle:.2f}', tuple(np.multiply(right_knee, [640, 480]).astype(int)), cv.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv.LINE_AA)

                    # Update stage and counters based on the selected exercise type
                    stage, counter, curl_counter, curl_stage = update_reps_and_curls(left_angle, stage, counter, curl_counter, curl_stage, exercise_type)
                
            except Exception as e:
                print(f"Error occurred: {e}")

            # Draw UI elements for exercise counter
            cv.rectangle(image, (0, 0), (225, 120), (245, 117, 16), -1)
            
            if exercise_type == 'press':
                cv.putText(image, 'PRESSES', (65, 12), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv.LINE_AA)
                cv.putText(image, str(counter), (60, 60), cv.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv.LINE_AA)
            
            elif exercise_type == 'curl':
                cv.putText(image, 'CURLS', (65, 12), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv.LINE_AA)
                cv.putText(image, str(curl_counter), (60, 60), cv.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv.LINE_AA)

            # Draw pose landmarks
            mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                     mp_drawing.DrawingSpec(color=(255, 255, 255), thickness=2, circle_radius=5),
                                     mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=3, circle_radius=5))

        cv.imshow("Pose and Exercise Detection AI", image)
        
        # Listen for keypress to select exercise
        key = cv.waitKey(1) & 0xFF
        if key == ord('1') and exercise_type is None:  # User selects Curls
            exercise_type = 'curl'
        elif key == ord('2') and exercise_type is None:  # User selects Presses
            exercise_type = 'press'

        if key == ord('q'):  # Exit the loop
            break

cap.release()
cv.destroyAllWindows()
```

```python
#Added Diet plan functionality
import cv2 as cv
import numpy as np
import mediapipe as mp

# Initialize MediaPipe Pose
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

# Calculate angle between 3 points
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    
    if angle > 180.0:
        angle = 360 - angle
    return angle

# Update reps (presses), curls, and squats based on angle
def update_reps_and_curls_and_squats(left_angle, right_angle, left_knee_angle, right_knee_angle, left_hip_angle, right_hip_angle, 
                                     stage, counter, curl_counter, curl_stage, squat_counter, squat_stage, exercise_type):
    # Logic for counting presses (reps)
    if exercise_type == 'press':
        if left_angle > 160 and right_angle > 160:  # Fully extended arm for presses
            stage = "down"
        if left_angle < 30 and right_angle < 30 and stage == "down":  # Fully contracted arm for presses
            stage = "up"
            counter += 1

    # Logic for counting curls (using left arm angle)
    if exercise_type == 'curl':
        if left_angle < 30:  # Arm fully contracted for curls
            curl_stage = "up"
        if left_angle > 140 and curl_stage == "up":  # Arm fully extended for curls
            curl_stage = "down"
            curl_counter += 1

    # Logic for counting squats (using knee and hip angles)
    thr = 165  # Threshold angle for squat detection
    if exercise_type == 'squat':
        # Squat down position: both knee and hip angles below threshold (i.e., angle < 165)
        if (left_knee_angle < thr) and (right_knee_angle < thr) and (left_hip_angle < thr) and (right_hip_angle < thr):
            squat_stage = "down"
        
        # Squat up position: both knee and hip angles above threshold (i.e., angle > 165)
        if (left_knee_angle > thr) and (right_knee_angle > thr) and (left_hip_angle > thr) and (right_hip_angle > thr) and (squat_stage == 'down'):
            squat_stage = 'up'
            squat_counter += 1
    
    return stage, counter, curl_counter, curl_stage, squat_counter, squat_stage

# Initialize camera and counters
cap = cv.VideoCapture(0)
counter = 0  # Counter for presses
curl_counter = 0  # Counter for curls
squat_counter = 0  # Counter for squats
stage = None
curl_stage = None
squat_stage = None
exercise_type = None  # Default exercise type is None

# BMI Calculation and Category
def calculate_bmi(weight, height):
    # Convert height from cm to meters for BMI calculation
    height_in_meters = height / 100.0  # Convert height to meters
    bmi = weight / (height_in_meters ** 2)
    return bmi

def get_bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

# Diet Plan Based on BMI Category
def get_diet_plan(bmi_category):
    if bmi_category == "Underweight":
        return ("Increase calorie intake with healthy foods. Include whole grains, lean proteins, dairy products, nuts, and avocados. "
                "Aim for 500-1000 extra calories per day.")
    elif bmi_category == "Normal weight":
        return ("Maintain a balanced diet with a mix of proteins, carbohydrates, and fats. Focus on whole foods, fruits, vegetables, "
                "and lean meats. Keep your calorie intake at maintenance level.")
    elif bmi_category == "Overweight":
        return ("Create a calorie deficit to lose weight. Include high-fiber foods like vegetables, fruits, whole grains, and lean proteins. "
                "Limit processed sugars and unhealthy fats. Regular exercise is essential.")
    else:
        return ("Consult a healthcare provider for personalized advice.")

# Display exercise options on the camera feed
def display_exercise_options(image):
    cv.putText(image, 'Select Exercise: ', (50, 50), cv.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv.LINE_AA)
    cv.putText(image, 'Curls [1]  Presses [2]  Squats [3]  BMI [4]', (50, 90), cv.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv.LINE_AA)

# Display BMI information and diet plan
def display_bmi_info(image, bmi, category, diet_plan):
    cv.putText(image, f'BMI: {bmi:.2f}', (50, 50), cv.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv.LINE_AA)
    cv.putText(image, f'Category: {category}', (50, 90), cv.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv.LINE_AA)
    cv.putText(image, f'Diet Plan:', (50, 150), cv.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv.LINE_AA)
    cv.putText(image, f'{diet_plan}', (50, 200), cv.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv.LINE_AA)

# Display input prompt for height and weight
def display_input_prompt(image, input_str, y_offset=150):
    cv.putText(image, f'Enter {input_str}: {height_input if input_str == "Height" else weight_input}', 
               (50, y_offset), cv.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv.LINE_AA)

# Handle user input for height and weight
def handle_keypress(key, input_str):
    global height_input, weight_input, exercise_type
    if key == ord('b'):  # Backspace to delete last character
        if input_str == "Height" and len(height_input) > 0:
            height_input = height_input[:-1]
        elif input_str == "Weight" and len(weight_input) > 0:
            weight_input = weight_input[:-1]
    elif key == 13:  # Enter key to confirm input
        if input_str == "Height":
            exercise_type = 'bmi_weight'  # Move to weight input
        elif input_str == "Weight":
            exercise_type = None  # Finish BMI input
            return True  # Allow exit after entering both height and weight
    elif key >= 48 and key <= 57:  # Digits 0-9
        if input_str == "Height":
            height_input += chr(key)
        elif input_str == "Weight":
            weight_input += chr(key)
    return False

# Initialize height and weight input variables
height_input = ""
weight_input = ""

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        image = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
        results = pose.process(image)
        image = cv.cvtColor(image, cv.COLOR_RGB2BGR)

        # If exercise type has not been selected, show the options on the screen
        if exercise_type is None:
            display_exercise_options(image)
        else:
            try:
                if results.pose_landmarks:
                    landmarks = results.pose_landmarks.landmark
                
                    # Left arm landmarks
                    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
                    left_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
                    left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
                
                    # Right arm landmarks
                    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
                    right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
                    right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

                    # Left leg landmarks (Hip, Knee, Ankle)
                    left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x, landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
                    left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
                    left_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
                    
                    # Right leg landmarks (Hip, Knee, Ankle)
                    right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
                    right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
                    right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]

                    # Calculate angles for arms
                    left_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
                    right_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)

                    # Calculate angles for knees
                    left_knee_angle = calculate_angle(left_hip, left_knee, left_ankle)
                    right_knee_angle = calculate_angle(right_hip, right_knee, right_ankle)

                    # Calculate angles for hips
                    left_hip_angle = calculate_angle(left_knee, left_hip, left_ankle)
                    right_hip_angle = calculate_angle(right_knee, right_hip, right_ankle)

                    # Update counters and stages based on the selected exercise
                    stage, counter, curl_counter, curl_stage, squat_counter, squat_stage = update_reps_and_curls_and_squats(
                        left_angle, right_angle, left_knee_angle, right_knee_angle, left_hip_angle, right_hip_angle,
                        stage, counter, curl_counter, curl_stage, squat_counter, squat_stage, exercise_type)
            
            except Exception as e:
                print(f"Error occurred: {e}")

            # Handle BMI input
            if exercise_type == 'bmi':
                display_input_prompt(image, "Height")
                if handle_keypress(cv.waitKey(1), "Height"):
                    exercise_type = 'bmi_weight'  # Move to weight input
            
            if exercise_type == 'bmi_weight':
                display_input_prompt(image, "Weight", 250)
                if handle_keypress(cv.waitKey(1), "Weight"):
                    # Debugging: Check input values
                    print(f"Height: {height_input}, Weight: {weight_input}")
                    
                    # Make sure height and weight are valid numbers before calculating BMI
                    if height_input and weight_input:
                        bmi = calculate_bmi(float(weight_input), float(height_input))  # Convert height to meters
                        category = get_bmi_category(bmi)
                        diet_plan = get_diet_plan(category)
                        
                        # Debugging: Check result
                        print(f"BMI: {bmi:.2f}, Category: {category}")
                        print(f"Diet Plan: {diet_plan}")
                        
                        # Display BMI info and diet plan on screen
                        display_bmi_info(image, bmi, category, diet_plan)
                    exercise_type = None  # Reset to allow selecting other exercises
        
        cv.imshow("Pose and Exercise Detection AI", image)

        # Handle keypress for exercise selection
        key = cv.waitKey(1) & 0xFF
        if key == ord('1') and exercise_type is None:  # User selects Curls
            exercise_type = 'curl'
        elif key == ord('2') and exercise_type is None:  # User selects Presses
            exercise_type = 'press'
        elif key == ord('3') and exercise_type is None:  # User selects Squats
            exercise_type = 'squat'
        elif key == ord('4') and exercise_type is None:  # User selects BMI
            exercise_type = 'bmi'

        if key == ord('q'):  # Exit the loop
            break

cap.release()
cv.destroyAllWindows()
```
